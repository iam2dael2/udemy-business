#### Drop Irrelevant Columns

```python
data["price_detail__currency"].unique()
```

```python
new_data = data.drop(["discount_price__amount", "discount_price__currency", "discount_price__price_string", 
                      "price_detail__currency", "is_paid", "is_wishlisted", "price_detail__price_string"], axis=1)
new_data.head()
```

#### Data Preprocessing

```python
new_data.info()
```

```python
# Check if avg_rating_recent is the same as rating
sum(new_data["avg_rating_recent"] == new_data["rating"]) == len(new_data)
```

```python
# Since it's True, then we drop avg_rating and avg_rating_recent
preprocessed_data = new_data.drop(["avg_rating", "avg_rating_recent"], axis=1)
preprocessed_data.head()
```

```python
preprocessed_data.info()
```

```python
# Convert "created" and "published_time" into datetime type
preprocessed_data["created"] = pd.to_datetime(preprocessed_data["created"], format="%Y-%m-%dT%H:%M:%SZ")
preprocessed_data["published_time"] = pd.to_datetime(preprocessed_data["published_time"], format="%Y-%m-%dT%H:%M:%SZ")

# Rename "price_detail__amount" into "price"
preprocessed_data.rename(columns={"price_detail__amount": "price"}, inplace=True)
```

```python
preprocessed_data["year"] = preprocessed_data["published_time"].dt.year
preprocessed_data.drop(["created", "published_time"], axis=1, inplace=True)
```

```python
preprocessed_data.head()
```

```python
preprocessed_data.shape
```

#### Answer Questions

```python
plt.style.use("seaborn-whitegrid")
```

#### 1. What is the most subscribed business course in Udemy?

```python
high_subs = preprocessed_data.sort_values(by=["num_subscribers"], ascending=True).tail(10)
high_subs
```

```python
high_subs.plot(x="title", y="num_subscribers", kind="barh", figsize=(10, 5))
plt.title("Title vs Num Subscribers", fontsize=16)
```

**Answer**: An Entire MBA in 1 Course: Award Winning Business School Prof is the most subscribed business course in Udemy, but any course whose rank below this rank are mostly used for data analyst or data science hard-skills.

#### 2. Which course that has the best overall rating?

```python
## Rating
best_rating = preprocessed_data.sort_values(by=["rating"], ascending=True)
best_rating.tail()
```

Five of these courses have the highest rating, **5.0**, but their number of reviews is much lower than any courses.

```python
## Rating with Num Reviews
best_overall_rating = preprocessed_data[["id", "title", "rating", "num_subscribers", "num_reviews", "year"]].copy()
best_overall_rating["overall_score"] = best_overall_rating["rating"] * best_overall_rating["num_reviews"]
best_10_rating = best_overall_rating.sort_values(by=["overall_score", "rating"], ascending=True).tail(10)
best_10_rating.tail()
```

```python
fig, axs = plt.subplots(1, 3)
fig.suptitle("Title vs")

axs[0].barh(best_10_rating["title"], best_10_rating["rating"])
axs[0].set(xlabel='Rating', ylabel='Title')

axs[1].barh(best_10_rating["title"], best_10_rating["num_reviews"])
axs[1].set(xlabel='Num Reviews', ylabel='Title')

axs[2].barh(best_10_rating["title"], best_10_rating["overall_score"])
axs[2].set(xlabel='Overall Score', ylabel='Title')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

**Answer**: The Complete SQL Bootcamp 2020: Go from Zero to Hero is the highest-rated business course in Udemy.

#### 3. Are higher priced-course have higher rating than lower priced-course?

```python
sns.displot(preprocessed_data["price"], kde=True)
```

```python
## Rating vs Price
sns.lmplot(x="price", y="rating", data=preprocessed_data, line_kws={'color': 'red'})
plt.title("Price vs Rating")
plt.show()
```

**Answer**: Price is not correlated well with rating.

#### 4. How is the correlation between among features?

```python
df_corr = preprocessed_data.copy()
df_corr = df_corr.drop("id", axis=1).corr()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df_corr, annot=True)
```

**Answer**: All of features are low-correlated among others, except "num_subscribers vs num_reviews". Num_subscribers is moderately high-correlated with num_reviews.

#### 5. How are the changes about courses among years?

```python
## Group by years and see total num_subscribers
rating_by_years = preprocessed_data.groupby("year")["num_subscribers"].sum()
rating_by_years
```

```python
rating_by_years.plot()

plt.title("year vs num_subscribers", fontsize=20, pad=20)
plt.xlabel("year", fontsize=12)
plt.ylabel("num_subscribers", fontsize=12)
```

**Answer**: Up to 2016, graph is moving upward to its peak with num_subscribers of *5.908.404*, then moving downward and ended in num_subscribers of *3e-6*

```python
## The most subscribed course for each year, then see graph
idxs = best_overall_rating.groupby(['year'])['overall_score'].transform(max) == best_overall_rating["overall_score"]
best_by_year = best_overall_rating[idxs].sort_values("year")
best_by_year.set_index("year")
```

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 6))
fig.suptitle("Most-Reviewed Business Course Each Year vs.", fontsize=18)

axs[0].plot(best_by_year["year"], best_by_year["overall_score"])
axs[0].set(xlabel='Year', ylabel='Overall Score')

axs[1].plot(best_by_year["year"], best_by_year["num_reviews"])
axs[1].set(xlabel='Year', ylabel='Num Reviews')

axs[2].plot(best_by_year["year"], best_by_year["num_subscribers"])
axs[2].set(xlabel='Year', ylabel='Num Subscribers')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

**Answer** : Up to 2020, SQL course published in 2016 is continously subscribed until then. However, other most-reviewed course in any years beside 2016 is stand still.
